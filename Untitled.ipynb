```python
#function prototype, inputs: list of twitter accounts links, search item, the time interval for another scrapping session,headers to be used in requests 
def scrape_twitter_accounts(twitter_accounts, ticker, time_interval, headers):
    with requests.Session() as session:
        session.headers.update(headers) # updates the headers
        while True:
            mentions_count = 0 #counter
            current_time = time.strftime("%H%M%S")
            
            for twitter_account in twitter_accounts: #loops the links in list
                response = session.get(twitter_account) #get request
                #print(response.status_code)
                if response.status_code == 200: #if response successful
                    soup = BeautifulSoup(response.text, 'html.parser') # text stored in soup variable
                    tweets = soup.find_all('div') #finds text of tweets
                    
                    for tweet in tweets:
                        print(tweet.text)
                        if ticker in tweet.text:
                            mentions_count += 1 #increase count when match found
            
            print(f"'{ticker}' was mentioned '{mentions_count}' times in the last '{time_interval}' minutes.")
            time.sleep(time_interval*60) #pause for time interval (in minutes)


headers = {#to act as a real user
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.9',
    'Referer': 'https://twitter.com/',
}

twitter_accounts = [ #list of twitter accounts
    'https://twitter.com/Mr_Derivatives',
    'https://twitter.com/warrior_0719',
    'https://twitter.com/ChartingProdigy',
    'https://twitter.com/allstarcharts',
    'https://twitter.com/yuriymatso',
    'https://twitter.com/TriggerTrades',
    'https://twitter.com/AdamMancini4',
    'https://twitter.com/CordovaTrades',
    'https://twitter.com/Barchart',
    'https://twitter.com/RoyLMattox'
]

ticker = '$SPX'
interval_minutes = 1

scrape_twitter_accounts(twitter_accounts, ticker, interval_minutes, headers)



```
